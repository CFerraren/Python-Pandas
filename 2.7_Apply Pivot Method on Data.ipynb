```python
# Import Libraries

import os
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# File Location

file = os.path.join('Resources', 'weather.csv')
```

```python
# Read csv to dataframe

df = pd.read_csv(file)
df
```

#### Pivot Function

```python
# Specify the index and columns

df.pivot(index='day', columns='city')
```

```python
# Specify the values 

df.pivot(index='day', columns='city', values='precipitation')
```

```python
# Specify the values 

df.pivot(index='day', columns='city', values='wind_speed')
```
