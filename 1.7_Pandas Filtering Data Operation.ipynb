```python
# Import Libraries

import os
import pandas as pd
```

```python
# Declare the file path

csvpath = os.path.join('Resources', 'imdb_1000.csv')
```

```python
# Read the csv file into a dataframe using read_csv function

df = pd.read_csv(csvpath)
df.head()
```

#### Applying filters on Dataframe

> 1. Using For Loop

  > Not a good implementation in real world problems as it takes more ressource and time to execute particularly for large data 

```python
# Get data with duration greater than 175

booleans = []

for length in df.duration:
    if length >= 175:
        booleans.append(True)
        
    else:
         booleans.append(False)
```

```python
df[booleans].head()
```

---

> 2. Using Series Implementation 

```python
df[df.duration >= 175][:10]
```
