```python
# Import Libraries

import os
import pandas as pd
```

```python
# Declare the file path

csvpath = os.path.join('Resources', 'imdb_1000.csv')
```

```python
# Read the csv file into a dataframe using read_csv function

df = pd.read_csv(csvpath)
df.head()
```

```python
#Sort the Series, default is ascending

df.duration.sort_values()[:10]
```

```python
df['duration'].sort_values(ascending = False)[:10]
```

```python
df['duration'].sort_values(ascending = True)[:10]
```

```python
# Sort the entire dataframe by the Duration column
#Ascending order

df.sort_values('duration')[:10]
```

```python
# Sort the entire dataframe by the Duration column
# Descending order

df.sort_values('duration', ascending=False)[:10]
```

#### Sorting Multiple Columns

```python
# Sort the entire dataframe by the Duration column
#Ascending order

df.sort_values(['genre','duration'])[:10]
```
