### Agenda: 




* **``fillna``** function to munge data


* **``ffill``** forward fill
    

* **``bfill``** forward fill


---

```python
# Import Libraries

import os
import pandas as pd
```

```python
# File Location

file = os.path.join('Resources', 'temperature.csv')
```

```python
# Read csv to dataframe

temp = pd.read_csv(file)
temp
```

```python
#Forward Filling the missing data

temp.fillna(method='ffill')
```

```python
#Backward Filling the missing data

temp.fillna(method='bfill')
```

```python
# Backward Filling the missing data
# Axis parameter
# axis =0 - based on rows or vertical substitution of data
# axis =1 - based on columnar or horizontal substitution of data

temp.fillna(method='bfill', axis=0)
```

```python
# Set limit parameter to limit the number of filling of data
# limit=1 - means that it will only backward fill up to 1, more than that like the next 
# missing willjust fill with NaN

temp.fillna(method='bfill', limit=1)
```
