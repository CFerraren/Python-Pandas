* **``fillna``** function to munge data

* Use of dictionary to fill-in different values for different columns

```python
# Import Libraries

import os
import pandas as pd
```

```python
# File Location

file = os.path.join('Resources', 'ufo.csv')
```

```python
# Read csv to dataframe

ufo = pd.read_csv(file)
ufo.head()
```

```python
# Fill all NaN with ``blank`` keyword in all dataframe

ufo_new = ufo.fillna('blank')
ufo_new[:20]
```

```python
# Fill all NaN with ``No Color`` for ``Colors Reported``
# Fill all NaN with ``No Shape`` for ``Shape Reported``
# Use dictionary to fill-in with different values for the different columns

ufo1 = ufo.fillna({'Colors Reported':'No Color', 'Shape Reported':'No Shape'})
ufo1[:20]
```
