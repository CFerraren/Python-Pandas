```python
# Import Libraries

import os
import pandas as pd
```

```python
# File Location

file = os.path.join('Resources', 'Products-1.xlsx')
```

```python
# Read csv to dataframe

dairy = pd.read_excel(file, header = [0,1,2])
dairy
```

#### Stack without Specifying the level of Stacking

```python
dairy.stack(level=[0,1])
```

```python
dairy1 = dairy.stack()
dairy1
```

```python
dairy1 = dairy.stack(level=[0,1,2])
dairy1
```

#### Unstack

```python
dairy1.unstack(level=[1,2])
```
