```python
# Import Libraries

import os
import pandas as pd
```

```python
# File Location

file = os.path.join('Resources', 'Products.xlsx')
```

```python
# Read csv to dataframe

dairy = pd.read_excel(file, header = [0,1])
dairy
```

#### Stack without Specifying the level of Stacking

```python
dairy1 = dairy.stack()
dairy1
```

```python
dairy1 = dairy1.stack()
dairy1
```

---

#### Stack Specifying the Level of Stacking

```python
# Stack the data using the level = 0 which is the State (California, Nevada)
dairy2 = dairy.stack(level=0)
dairy2
```

```python
# Stack the data using the level = 0 which is the State (California, Nevada)
dairy3 = dairy.stack(level=[0,1])
dairy3
```

---

```python
dairy1.unstack()
```

```python
dairy1.unstack(level=0)
```

```python
dairy1.unstack(level=1)
```

```python
dairy1.unstack(level=[0,2])
```

```python
dairy2.unstack(level=1)
```

```python
dairy3.unstack(level=0)
```
