### Agenda: 




* **``to_excel``** saving data to excel file


* **``ExcelWriter``** saving multi-dataframe to excel file




---

```python
# Import Libraries

import os
import pandas as pd
```

```python
# File Location

file = os.path.join('Resources', 'weather-1.csv')
```

```python
# Read csv to dataframe

df = pd.read_csv(file)
df
```

```python
df1 = df.copy()
```

> #### Replace all Pasadena with Glendale

```python
# User replace

df1.replace('Pasadena', 'Glendale', inplace=True)

df1
```


> #### Store our updated data into a Excel file

```python
#Save data into an excel format file


df1.to_excel('Resources/weather_excel.xlsx')
```

```python
# Save data into an excel format file
# Name the worksheet to ``Weather``
# Not including the index rows

df1.to_excel('Resources/weather_excel_1.xlsx', sheet_name='Weather', index=False)
```

```python
# Specify the row and column where the table will be saved

df1.to_excel('Resources/weather_excel_2.xlsx', sheet_name='Weather', index=False, startrow=3, startcol=4)
```

---

#### Saving 2 dataframe into an excel file

```python
df.tail()
```

```python
df1.tail()
```

```python
with pd.ExcelWriter('Resources/weather_excel_combined.xlsx') as writer:
    df.to_excel(writer, sheet_name='Weather_Pasadena', index=False)
    df1.to_excel(writer, sheet_name='Weather_Glendale', index=False)
```
