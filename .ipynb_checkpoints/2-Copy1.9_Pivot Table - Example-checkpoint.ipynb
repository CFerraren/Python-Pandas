```python
# Import Libraries

import os
import pandas as pd
from pandas import Grouper
```

```python
# File Location

file = os.path.join('Resources', 'weather-1.csv')
```

```python
# Read csv to dataframe

df = pd.read_csv(file)
df
```

> #### Replace all Pasadena with Glendale

```python
# User replace

df.replace('Pasadena', 'Glendale', inplace=True)

df
```


> #### Store our updated data into a csv file

```python
#this copy will include the index series into a csv


df.to_csv('Resources/weather_updated')
```

```python
# Excluding the index into a csv file
# Specify the index= False

df.to_csv('Resources/weather_updated_no_Index', index=False)
```

```python
# Excluding the index into a csv file
# Specify the index= False


# Excluding the Table Header into a csv file
# Specify the header= False

df.to_csv('Resources/weather_updated_no_Header', index=False, header=False)
```

> #### Only Specific Columsn to Copy and not the entire table

```python
# Excluding the Table Header into a csv file
# Specify the header= False

df.to_csv('Resources/weather_updated_spcific_columns', index=False, columns=['day','city', 'wind_speed'])
```
