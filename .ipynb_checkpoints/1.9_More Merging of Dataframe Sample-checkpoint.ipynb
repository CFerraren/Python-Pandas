### Agenda: 


* Merging dataframe


* Inner, Outer, Left, Outer


* **``Indicator``** to have details which data comes from what dataframe
    
    


---

```python
# Import Libraries

import pandas as pd
```

```python
#Create Dataframe

df1 = pd.DataFrame([['CA', 'Sacramento'], ['TX', 'Austin'],['AK', 'Juneau'], 
                    ['KY', 'Frankfort'], ['AZ', 'Phoenix']],
                  columns=['State', 'Capital'])

df1
```

```python
#Create Dataframe

df2 = pd.DataFrame([['CA', 567000000], ['TX', 545000000],['AK', 148000000], 
                    ['KY', 117000000], ['WY', 311000000]],
                  columns=['State', 'search_results'])

df2
```

```python
#Inner Merge

pd.merge(df1,df2, how='inner', on='State')
```

```python
#Outer Merge

pd.merge(df1,df2, how='outer', on='State')
```

```python
#Left Merge

pd.merge(df1,df2, how='left', on='State')
```

```python
#Right Merge

pd.merge(df1,df2, how='right', on='State')
```

```python
#outer Merge with Indicator to True

pd.merge(df2,df2, how='outer', on='State', indicator=True)
```
