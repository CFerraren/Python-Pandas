* **``Melt``** function to munge data
    
    


---

```python
# Import Libraries

import pandas as pd
```

```python
#Create Dataframe

df1 = pd.DataFrame({'days':['Mon','Tue', 'Wed', 'Thu', 'Fri','Sat','Sun'],
                   'temp':[33,34,31,34,35,37,34],
                   'humidity':[60,68,55,57,45,48,55]},
                  columns=['days', 'temp','humidity'])

df1
```

#### Applying Melt

```python
pd.melt(df1, id_vars='days')
```

```python
#Specify the variable name and value name

pd.melt(df1, id_vars='days', var_name='metric', value_name='measurement')
```
