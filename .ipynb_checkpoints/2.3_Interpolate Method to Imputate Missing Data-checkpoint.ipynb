```python
# Import Libraries

import os
import pandas as pd
```

```python
# File Location

file = os.path.join('Resources', 'temperature.csv')
```

```python
# Read csv to dataframe

temp = pd.read_csv(file)
temp
```

```python
#Check Datatypes

temp.dtypes
```

```python
#Change the Format of the Day to datetime format

temp['Day'] = pd.to_datetime(temp.Day)
```

```python
#Check Datatypes

temp.dtypes
```

```python
temp.set_index(temp.Day, inplace=True)
```

```python
temp.drop('Day', axis=1)
```

```python
# To set method to ``time``
# 1st: Must set the index to the date
# 2nd: Must set the data type to datetime format


temp.interpolate(method='time')
```
