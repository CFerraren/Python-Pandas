### Agenda: 




* **``crosstab``** function to analyze variables from another target variable in the dataset

* for it to work, data can be aggregation or have aggregation values like the ``Name`` that can be used.

* use Unique function to support the nuber of unique values





---

```python
# Import Libraries

import os
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# File Location

file = os.path.join('Resources', 'titanic_train.csv')
```

```python
# Read csv to dataframe

titanic = pd.read_csv(file)
titanic.head()
```

#### Crosstab Function

```python
# Needs two variable 

pd.crosstab(titanic.Pclass, titanic.Survived)
```

```python
# Needs two variable and plot

pd.crosstab(titanic.Pclass, titanic.Survived).plot(kind='bar')
```

#### Example of Crosstab not making sense

```python
# Name have too many unique values and no repeating values
# Doesn't make sense to use crosstab

pd.crosstab(titanic.Name, titanic.Survived).head()
```

```python
# margins parameter to include the Total

pd.crosstab(titanic.Embarked, titanic.Survived, margins=True)
```

```python
pd.crosstab(titanic.Embarked, titanic.Survived).plot(kind='bar')
```

```python
pd.crosstab(titanic.Sex, titanic.Survived)
```

```python
pd.crosstab(titanic.Sex, titanic.Survived).plot(kind='bar')
```

```python
#Combining using a List

pd.crosstab([titanic.Pclass,titanic.Sex,titanic.Embarked], titanic.Survived, margins=True)
```

```python
#Combining using a List

pd.crosstab([titanic.Survived,titanic.Pclass,titanic.Sex],titanic.Embarked,  margins=True)
```

```python
#normalize to the whole sample set

pd.crosstab(titanic.Embarked, titanic.Survived, normalize=True)
```

```python
#normalize across the column

pd.crosstab(titanic.Embarked, titanic.Survived, normalize='columns')
```

```python
#normalize across the row

pd.crosstab(titanic.Embarked, titanic.Survived, normalize='index')
```

```python
pd.crosstab(titanic.Pclass, titanic.Survived, values=titanic.Age, aggfunc=['mean','max','min'])
```

```python
pd.crosstab(titanic.Embarked, titanic.Survived, values=titanic.Age, aggfunc=['mean','max','min'])
```
