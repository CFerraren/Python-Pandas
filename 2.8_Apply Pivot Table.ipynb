```python
# Import Libraries

import os
import pandas as pd
```

```python
# File Location

file = os.path.join('Resources', 'weather-1.csv')
```

```python
# Read csv to dataframe

df = pd.read_csv(file)
df
```

#### Using Pivot Method

> using Pivot method with duplicated data (index) will yield an error

```python
# Specify the values 

df.pivot(index='day', columns='city')
```

#### Using Pivot Table Method

 > no error and it will aggregate base

```python
# Specify the index and columns

df.pivot_table(index='day', columns='city')
```

```python
# Specify the values 

df.pivot_table(index='day', columns='city', values='wind_speed')
```

```python
# Specify the values 

df.pivot_table(index='day', columns='city', values='wind_speed', aggfunc=['min', 'max', 'mean', 'std'])
```
